## Import Required modules

```python
import pandas as pd
import numpy as np
```

# Create DataFrames

```python
df1 = pd.DataFrame({
    'Date':['2013-11-24','2013-11-24','2013-11-24','2013-11-24'],
    'Fruit':['Banana','Orange','Apple','Celery'],
    'Num':[22.1,8.6,7.6,10.2],
    'Color':['Yellow','Orange','Green','Green'],
    })

df1
```

```python
df2 = pd.DataFrame({
    'Date':['2013-11-24','2013-11-24','2013-11-24','2013-11-24','2013-11-25','2013-11-25'],
    'Fruit':['Banana','Orange','Apple','Celery','Apple','Orange'],
    'Num':[22.1,8.6,7.6,10.2,22.1,8.6],
    'Color':['Yellow','Orange','Green','Green','Red','Orange'],
    })

df2
```

# Approch 01

```python
df = pd.concat([df1,df2])  # https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html
df
```

```python
df.drop_duplicates()
```

```python
df.drop_duplicates(keep=False)
```

# Approach 2

```python
df = pd.concat([df1, df2]) # concatenate dataframes 
df
```

```python
df = df.reset_index(drop=True) # Reset Index
df
```

```python
df_group = df.groupby(list(df.columns))
df_group
```

```python
df_group.aggregate(np.sum)
```

```python
df_group.apply(lambda x: x) 
```

```python
df_group.groups.values()
```

```python
# get index of unique records
idx = [x[0] for x in df_group.groups.values() if len(x) == 1]  
idx
```

```python
df.reindex(idx)
```
