# Database connectivity and operations using Python.

```
# import required modules
import sqlite3

#connecting with the database.
db = sqlite3.connect("my_database.db")

# Drop table if it already exist using execute() method.
db.execute("drop table if exists customers")

# Create table as per requirement
db.execute("create table customers(id int, name text, age int, city string)")

#inserting values inside the created table
db.execute("insert into customers(id, name, age, city) values(101, 'Teja', 7, 'Bangalore')")
db.execute("insert into customers(id, name, age, city) values(102, 'Bhavishya',3 ,'Bangalore' )")
db.execute("insert into customers(id, name, age, city) values(103, 'Sagar', 10 ,'Yemmiganur' )")
db.execute("insert into customers(id, name, age, city) values(104, 'Satish', 32,  'Bangalore' )")
db.execute("insert into customers(id, name, age, city) values(104, 'Ramya', 27,  'Bangalore' )")

# committing the current chnages
db.commit()
```

```
results = db.execute("select * from customers order by id")
for row in results:
    print(row)
```

Note: result will always returns in Tuples

```
results = db.execute("select * from customers where age >= 10 ")
for row in results:
    print(row)
```

```
results = db.execute("select name, age from customers order by age desc ")
for row in results:
    print(row)
```
